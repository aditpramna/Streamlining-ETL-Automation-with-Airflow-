```python
import pandas as pd
import great_expectations as gx

from great_expectations.checkpoint import Checkpoint
```

# **BUILD CONTEXT**

```python
context = gx.get_context()
```

# **BUILD DATA SOURCE**

```python
df = pd.read_csv('P2M3_Aditya_Pramana_data_clean.csv')
```

```python
datasource = context.sources.add_pandas(name="pandas_datasource")

name = "Breast Cancer"
data_asset = datasource.add_dataframe_asset(name=name)

my_batch_request = data_asset.build_batch_request(dataframe=df)
```

# **BUILD EXPECTATION SUITE AND VALIDATOR**

```python
# created an Expectation Suite
context.add_or_update_expectation_suite("my_expectation_suite")

# create a Validator using above Expectation Suite
validator = context.get_validator(
    batch_request=my_batch_request,
    expectation_suite_name="my_expectation_suite",
)

validator.head()
```

```python
# Expectation 1: Dataset Should Not Be Empty
validator.expect_table_row_count_to_be_between(min_value=1, max_value=None)
```

```python
# 2. Expectation3:  untuk unik
validator.expect_column_values_to_be_unique("id")
```

```python
# Expectation 3: 'diagnosis' column only contains the values 1 and 0
validator.expect_column_values_to_be_in_set(column='diagnosis', value_set=['M', 'B'])
```

```python
# Expectation 4: 'diagosis' to be of object type
validator.expect_column_values_to_be_of_type(column='diagnosis', type_='object')
```

```python
# Expectation 5: 'compactness_mean' Should Not Contain NULL Values

validator.expect_column_values_to_not_be_null(column='compactness_mean')
```

```python
# Expectation 6: 'radius_mean' Should Have a Maximum Value Below a Certain Threshold

validator.expect_column_max_to_be_between(column='radius_mean', min_value=6.981, max_value=28.11)
```

```python
# Expectation 7: 'perimeter_mean' Should Have a Minimum Value Above a Certain Threshold

validator.expect_column_min_to_be_between(column='perimeter_mean', min_value=43.79, max_value=188.5)
```

```python
# Save validator into expectation suite
validator.save_expectation_suite(discard_failed_expectations=False)
```

# **BUILD CHECKPOINT**

```python
checkpoint = context.add_or_update_checkpoint(
    name="my_checkpoint",
    validations=[
        {
            "batch_request": my_batch_request,
            "expectation_suite_name": "my_expectation_suite",
        },
    ],
)
```

```python
checkpoint_result = checkpoint.run()
```
